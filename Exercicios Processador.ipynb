O código de operação é formado por 8 bits, então no total há 2^8=256 estilos de Instruções, como há 3 tipos de instrução então será 256/3= 85 instruções para cada tipo.
O tipo A (Da célula 0 até a 84) contém 16 bits, dos quais 8 são do código de operação e os outros 8 são de operando, como cada célula é 1 byte (8 bits) então serão necessários 2 célas para uma instrução, então o endereço do PC (CI) deve sofrer acréscimos de 2 em 2, ou seja:   PC <- PC + 2

O tipo B (Da célula 85 até a 170) são 32 bits, o que equivale a 4 células por instrução, então o PC deve atualizar de 4 em 4, ou seja: PC <- PC + 4

O tipo C (Da célula 171 até a 256) são 48 bits, o que equivale a 6 células por instrução, então o PC deve atualizar de 6 em 6, ou seja: PC <- PC + 6
